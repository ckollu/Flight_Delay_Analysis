```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
print('imported')
# Step 1: Load CSV and clean blanks
df = pd.read_csv('flights.csv', dtype=str)  # Read all as strings to catch blanks
print('loaded')
df = df.applymap(lambda x: None if pd.isna(x) or str(x).strip() == '' else x)
print('replaced')
# Optional: convert numeric columns back to float/int if needed
# df['some_column'] = df['some_column'].astype(float)

# Step 2: Create SQLAlchemy engine
engine = create_engine('mysql+pymysql://root:Chandra%402001@localhost/aviation_analysis')
print('connected')
# Step 3: Push to MySQL
df.to_sql('raw_flights', con=engine, if_exists='replace', index=False)

print("✅ Data pushed to MySQL with NULLs for empty cells.")
```
