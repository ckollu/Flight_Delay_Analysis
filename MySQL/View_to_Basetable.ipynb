```python
import mysql.connector
import time
```

```python
def batch_transfer_data():
    conn = mysql.connector.connect(
        host='localhost',
        user='root',
        password='Chandra@2001',
        database='aviation_analysis'
    )
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS cleaned_flight_data_table AS SELECT * FROM cleaned_flight_data_optimized WHERE 1=0")
        # Get total count
    cursor.execute("SELECT COUNT(*) FROM cleaned_flight_data_optimized")
    total_count = cursor.fetchone()[0]
    
    batch_size = 10000
    offset = 0
    
    while offset < total_count:
        try:
            query = f"""
            INSERT INTO cleaned_flight_data_table 
            SELECT * FROM cleaned_flight_data_optimized 
            LIMIT {offset}, {batch_size}
            """
            cursor.execute(query)
            conn.commit()
            
            print(f"Transferred {offset + batch_size if offset + batch_size < total_count else total_count} of {total_count} records")
            
            offset += batch_size
            time.sleep(0.5)  # Small delay
            
        except Exception as e:
            print(f"Error at offset {offset}: {e}")
            time.sleep(5)  # Wait longer if error occurs
            continue
    
    cursor.close()
    conn.close()
if __name__ == "__main__":
        batch_transfer_data()
```
